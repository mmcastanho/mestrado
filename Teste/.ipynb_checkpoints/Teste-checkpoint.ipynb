```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import xml.etree.ElementTree as ET
```

```python
file = open("Consultas.txt" ,"r", encoding='latin-1')
```

```python
text3 = file.read().replace('\n', '')
```

```python
print(text3)
```

```python
text4 = "<tag>"+text3+"</tag>"
```

```python
text4
```

```python
root = ET.fromstring(text4)
lista = list()
for top in root.findall('top'):
    title = top.find('PT-title').text
    desc = top.find('PT-desc').text
    num = top.find('num').text
    lista.append((num,title,desc))
```

```python
df = pd.DataFrame(lista, columns=['num', 'title', 'desc'])
```

```python
df
```

```python
df2 = pd.DataFrame(list(range(0,100)), columns=['result'])
```
